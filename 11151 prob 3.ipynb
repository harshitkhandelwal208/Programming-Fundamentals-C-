```python
L = input("Enter the elements of the list, seperated by a space :-")
List = [ int(i) for i in L.split()]
List.sort()
num = List[0]
for i in List:
    if(num == i):
        if(i != List[0]):
            List.remove(i)
    num = i
print(List)
```

```python
L1 = input("Enter the elements of the first list, seperated by a space :-\t").split()
L2 = input("Enter the elements of the second list, seperated by a space :-\t").split()
if (len(L1)>len(L2)):
    length = len(L1)
else:
    length = len(L2)
isSimilar = False
for i in range(length):
    if(L1[i]==L2[i]):
        isSimilar = True
        break
if(isSimilar):
    print("The two lists have atleast one common member")
else:
    print("The two lists have no common member")
```
