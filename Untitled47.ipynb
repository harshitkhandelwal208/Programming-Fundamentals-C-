```python
L = input("Enter the elements of the list, seperated by a space :-")
List = [ int(i) for i in L.split()]
odd = 0
even = 0
for i in List:
    if(i%2==0):
        even = even + 1
    else:
        odd = odd + 1
print("Odd :-", odd, "\nEven :-", even)
```

```python
import math
start = int(input("Enter the starting value of the range :- "))
end = int(input("Enter the ending value of the range :- "))
def isPrime(n):
    if n < 2:
        return False
    for i in range(2,int(math.sqrt(n))+1):
        if n % i == 0:
            return False
    return True
for i in range(start,end+1):
    if isPrime(i) == False:
        print(i)
```

```python
List = input("Enter the elements of the list(seperated by spaces) :- ").split()
digits = 0
alphabets = 0
for i in List:
    if i.isdigit():
        digits = digits + 1
    else:
        alphabets = alphabets+1
print("Digits :-", digits, "\nAlphabets :-", alphabets)
```
