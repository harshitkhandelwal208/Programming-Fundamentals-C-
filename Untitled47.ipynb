```python
L = input("Enter the elements of the list, seperated by a space :-")
List = [ int(i) for i in L.split()]
odd = 0
even = 0
for i in List:
    if(i%2==0):
        even = even + 1
    else:
        odd = odd + 1
print("Odd :-", odd, "\nEven :-", even)
```

```python
import math
start = int(input("Enter the starting value of the range :- "))
end = int(input("Enter the ending value of the range :- "))
def isPrime(n):
    if n < 2:
        return False
    for i in range(2,int(math.sqrt(n))+1):
        if n % i == 0:
            return False
    return True
for i in range(start,end+1):
    if isPrime(i) == False:
        print(i)
```

```python
s1 = input("Enter side/angle1 :- ")
s2 = input("Enter side/angle2 :- ")
s3 = input("Enter side/angle3 :- ")
if s1 == s2 == s3:
    print("The triangle is an equilateral triangle")
elif s1==s2 or s1==s3 or s2==s3:
    print("The triangle is an isosceles triangle")
else:
    print("The triangle is a scalene triangle")
```

```python
choice = int(input("Enter your choice:- \n(1) Decimal to binary and hexadecimal\n(2) Binary to decimal and hexadecimal\n(3) Hexadecimal to decimal  and binary"))
if choice == 1:
    a = int(input("Enter a decimal number :-"))
    b = bin(a)
    c = hex(a)
    print("Binary :- ", b, "Decimal :- ", a, "Hexadecimal :- ", c)
elif choice == 2:
    b = bin(input("Enter a binary number :-"))
    a = int(b)
    c = hex(a)
    print("Binary :- ", b, "Decimal :- ", a, "Hexadecimal :- ", c)
elif choice==3:
    c = hex(input("Enter a hexadecimal number :-"))
    a = int(c)
    b = bin(c)
    print("Binary :- ", b, "Decimal :- ", a, "Hexadecimal :- ", c)
```
